```python
google_df = pd.DataFrame(requests.get("https://iex.lewagon.com/stable/stock/googl/chart/3m").json())
google_df['company_code'] = 'google'
google_df['date'] = pd.to_datetime(google_df['date'], format="%Y-%m-%d")
google_df.set_index(keys='date', inplace = True)
google_df.head(5)
```

```python
apple_df = pd.DataFrame(requests.get("https://iex.lewagon.com/stable/stock/aapl/chart/3m").json())
apple_df['company_code'] = 'apple'
apple_df['date'] = pd.to_datetime(apple_df['date'], format="%Y-%m-%d")
apple_df.set_index(keys='date', inplace = True)
apple_df.head(5)
```

```python
meta_df = pd.DataFrame(requests.get("https://iex.lewagon.com/stable/stock/meta/chart/3m").json())
meta_df['company_code'] = 'meta'
meta_df['date'] = pd.to_datetime(meta_df['date'], format="%Y-%m-%d")
meta_df.set_index(keys='date', inplace = True)
meta_df.head(5)
```

```python
amazon_df = pd.DataFrame(requests.get("https://iex.lewagon.com/stable/stock/amzn/chart/3m").json())
amazon_df['company_code'] = 'amazon'
amazon_df['date'] = pd.to_datetime(amazon_df['date'], format="%Y-%m-%d")
amazon_df.set_index(keys='date', inplace = True)
amazon_df.head(5)
```

```python
all_stock_df1 = pd.concat([google_df, apple_df, meta_df, amazon_df])
all_stock_df1
```

```python
# Pivot the table to get company_code as columns
pivot_df = all_stock_df1.pivot(columns='company_code', values='close')
pivot_df.plot(kind='line')
```

```python
pivot_df
```

```python
def normalize(series):
    return series/series[0]

normalized_pivot = pivot_df.apply(normalize, axis=0)
```

```python
normalized_pivot.plot(kind='line')
```
