```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import requests

url = "https://iex.lewagon.com/stable/stock/aapl/chart/3m"
api_data = requests.get(url).json()
```

```python
apple_df = pd.DataFrame(api_data)
apple_df.head(5)
```

```python
apple_df.info()
```

__Convert the date column to a datetime object__

```python
apple_df['date'] = pd.to_datetime(apple_df['date'], format="%Y-%m-%d")
#apple_df.info()
```

__Set the date column as the index__

```python
apple_df.set_index('date')
```

__Let's plot with 'close'__

```python
apple_df['close'].plot()
```

```python
apple_df[['open', 'close', 'high', 'low']].plot(figsize=(16,4))
```
